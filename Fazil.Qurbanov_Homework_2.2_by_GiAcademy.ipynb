```python
# Homework 2.2
'''
Question 1
Create a function containing the student's name 
and admission score and call that function.

'''
def call_name_and_score(st_name,st_score ): 
    name_list = []
    score_list = []
    new_dict = {}    
    name_list.append(st_name)
    score_list.append(st_score)
    
    for _ in range(len(name_list)):
        new_dict[st_name] = st_score
    print(new_dict)
    print(f"Name : {st_name} \nScore: {st_score}")

call_name_and_score("B", 25)  

#============================================================
'''
Question 2
Create a calculation() function that takes two variables 
and calculates their product.

'''
def calc_product(count , price):
    product = count * price
    print(product)
calc_product(15, 20)

#============================================================
'''
Question 3
Create a list of up to 20 even numbers.

'''
def create_20_even_numbers():
    count = 0
    even_list = []
    for i in range(2, 200, 6):
        if i%2==0:
            even_list.append(i)
            count+=1
            if count==20:
                break
    print(even_list)
create_20_even_numbers()
#============================================================
'''
Question 4
Using map and lambda functions, construct a function that multiplies 
each element in the list numbers = [23, 32, 12, 5, 14, 76, 90, 56] by 10 and 
returns the result.

'''
numbers = [23, 32, 12, 5, 14, 76, 90, 56]
new_list = list(map(lambda x: x*10 , numbers))
print(new_list)

#============================================================
'''
Question 5
Write a function that counts consonants in a word

'''
str1 = input("enter string : ")
consonants = "aıeiəoöuü"

count = 0
for i in range(len(str1)):
    if str1[i].lower() in consonants and str1[i].isalpha():
        count+=1
    
print(count)

#============================================================
'''
Question 6
Write a program that displays the numbers less than 50 from the list 
numbers = [23, 32, 12, 5, 14, 76, 90, 56] using matching functions.

'''
numbers = [23, 32, 12, 5, 14, 76, 90, 56]
for i in numbers[:]:    
    if i>=50:
        numbers.remove(i)    
print(numbers)

#============================================================
'''
Question 7
Write a program that calculates the 5th power root of an input 
number using the pow() function.

'''
def calc_power_root():
    num = float(input("enter any number for calculate 5th power root :"))
    root = pow(num , 1/5)
    print(root)
calc_power_root()

#============================================================
'''
Question 8
Write a program that calculates the largest element in the list 
numbers = [100, 202, 12, 35, 87, 56, 90, 65].
Note: lambda, reduce

'''
from functools import reduce
numbers = [-50,100, 202,-12, 12, 35, 87, 565, 90, 65]

largest = reduce(lambda a, b: a if a > b else b,  numbers)
print(largest)

#============================================================

'''
Question 9
Write a program that calculates the sum of all elements in the list 
numbers = [78, 34, 90, 45, 23, 19, 97, 56].

'''
numbers = [78, 34, 90, 45, 23, 19, 97, 56]
sum = reduce(lambda x, y: x+y, numbers )
print(sum)

#============================================================

'''
Question 10
Write a program that calculates the square of an input number using the appropriate module.

'''
import math 
def square_num(num, sq=2):
    num = float(input("enter numbers : "))
    print(math.pow(num,sq))
square_num(15)
    
```
